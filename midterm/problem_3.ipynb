```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# Define the transfer function coefficients
num = [4, 0, 11, 0, 9]
den = [16, 16, 64, 46, 60, 18, 0]

# Create the system
G = ct.TransferFunction(num, den)

# Step Response
t, y = ct.step_response(G)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(t, y, color='#007B8A', linewidth=2)
plt.title('Step Response (Node x2)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)

# Pole-Zero Map
plt.subplot(1, 2, 2)
ct.pzmap(G)
plt.title('Pole-Zero Map')
plt.tight_layout()
plt.show()

# Print poles for analysis
print("Poles:", G.poles())
print("Zeros:", G.zeros())
```
