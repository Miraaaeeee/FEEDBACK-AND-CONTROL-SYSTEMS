```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Define Transfer Function: G(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
num = [1]
den = [1, 2, 3, 2, 0]
sys = ctrl.TransferFunction(num, den)

# Calculate Poles
poles = ctrl.poles(sys)
print(f"System Poles: {poles}")

# 1. Step Response Analysis
t, y = ctrl.step_response(sys)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, y, label='Output θ₂(t)', color='#007B8A', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.legend()

# 2. Pole-Zero Map
plt.subplot(1, 2, 2)
ctrl.pzmap(sys, plot=True)
plt.title('Pole-Zero Map')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
