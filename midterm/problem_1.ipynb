```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# Define Transfer Function G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = ct.TransferFunction(num, den)

# 1. Pole-Zero Map
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
ct.pzmap(sys, plot=True)
plt.title('Pole-Zero Map')
plt.grid(True)

# 2. Step Response
t, y = ct.step_response(sys)
plt.subplot(1, 2, 2)
plt.plot(t, y, 'b', lw=2)
plt.title('System Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# Display Poles
print("System Poles:", ct.poles(sys))
```
